# INDEX MANAGEMENT
- creating a new index column **reset_index()**
- names the un-named column **reset_index(name='give the name')**
- selecting which col to be the index column **set_index()**
- merging to dataframes **merge()**
- appending a dataframe below or onto another **pd.concat()**
- exporting data **to_csv(), to excel(), to_clipboard(), etc**

### reset_index()

```python
df_usa = df[df['job_country'] == 'United States']
df_usa
```

```python
# this creates an index colunm
df_usa.reset_index(inplace=True)
df_usa
```

### set_index

```python
# this makes the selected column to be the index in the table however it deletes the other index columns.
df_usa.set_index('job_index', inplace=True)
df_usa
```

```python
median_pivot = df_usa.pivot_table(index='job_title_short', values='salary_year_avg', aggfunc=['median', 'min', 'max'])
median_pivot
```

```python
# sorting
median_pivot.sort_values(by=('median', 'salary_year_avg'), ascending=False)
```

```python
# this can also help when there is no index column header
median_pivot.sort_index()
```

## Exercise
- make a line graph of job count against month for 'data Analyst, data scientist and data engineer

```python
df_exe = df.copy()
```

```python
# testing
id(df_exe) == id(df)
```

```python
# making a month column
df_exe['job_posted_month'] =df_exe['job_posted_date'].dt.strftime('%B')

```

```python
# make a pivot table 
count_pivot = df_exe.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
count_pivot.reset_index(inplace=True) # reset index so that job_month can be viewed as a column
# Assign month names to month numbers so as to easily sort from first month to last
count_pivot['job_month_no'] = pd.to_datetime(count_pivot['job_posted_month'], format='%B').dt.month
count_pivot.sort_values('job_month_no', inplace=True)
count_pivot.set_index('job_posted_month', inplace=True) # put back job_month as index 
count_pivot.drop(columns='job_month_no', inplace=True) # remove the month_no after serving it purpose
jobs = ['Data Analyst', 'Data Engineer', 'Data Scientist']
count_pivot = count_pivot[jobs]
count_pivot
count_pivot.plot(kind='line')
```

# merging data frames

```python
df_software = pd.read_csv('https://lukeb.co/software_csv')
df_software
```

```python
# set the job month as the index to use as a join to mearge the two data feames
df_software.set_index('job_posted_month', inplace=True)
```

```python
df_jobs = count_pivot.merge(df_software, on="job_posted_month")
top5 = df_jobs.sum().sort_values(ascending=False).head(5).index.to_list()
df_jobs[top5].plot(kind='line')
plt.ylim(0,30000)
```

# concat
- this is a pandas function

```python
# make some dummy dataframes to learn concat
df_cc = df_copy1
df_cc['job_posted_month'] = df_cc['job_posted_date'].dt.strftime('%b')
```

```python
months = df_cc['job_posted_month'].unique()
months
```

```python
# created data frames for each month using dictionaries
dict_months = {month: df_cc[df_cc['job_posted_month'] == month] for month in months}
dict_months['Jan']
```

```python
# using concat
dt_q1 = pd.concat([dict_months['Jan'],dict_months['Feb'], dict_months['Mar']], ignore_index=True)
dt_q1['job_posted_month'].value_counts().plot(kind='bar')
```
