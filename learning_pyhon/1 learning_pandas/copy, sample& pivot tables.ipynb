## the sample, copy methods and pivot tables
- df.**sample('number of rows to return', *random_state*='asign a refernce number for this sample made')** sample method returns a sample of the values in the dataframe. 

- df.**copy()** copy makes a copy of the data frame
- df.**pivot_table(*values*='salary_year_avg', *index*='job_country', *columns*='job_title_short', *aggfunc*='median')**
- **df.loc[]** loc returns a df with data from the specified characters in a column

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('C:/Users/Ian/Desktop/panda ex/newdata_jobs3.csv')
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# sample method resturns a sample of the values in the dataframe.
df.sample(10)
df.sample(5, random_state=1)
# random_state returns the same sample every single time.
# random_state is a way to asign a number to a sample.
```

```python
# when you call out the random_state asign it brings back the sample 
# where that random_state was a signed
df.sample(5, random_state=1)
```

```python
# copy method makes a copy of data
df_copy = df.copy()
```

```python
id(df_copy) == id(df)
```

### pivot tables

```python
# note; size method : counts all rows (column/value does not to be specified)
# count method: counts non_Na(non_null) values (column/value needed to be specified)
# pivot tables are just like in excel
df.pivot_table(index='job_title_short', aggfunc='size') 
# groupby method cal also work df.groupby(['job_title_short']).size()
```

```python
df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short', aggfunc='median').head(5)
```

```python
# making a bar graph of salary for (data analyst, enginner and scientist) againt top 6 countries
top_countries = df['job_country'].value_counts().head(6).index # taking the top 6 countries
pt = df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short', aggfunc='median')
pt = pt.loc[top_countries]
jobs = ['Data Analyst', 'Data Engineer', 'Data Scientist']
pt = pt[jobs].sort_values(by=['Data Analyst'], ascending=False)
pt.plot(kind='bar')
pt[jobs]
```
