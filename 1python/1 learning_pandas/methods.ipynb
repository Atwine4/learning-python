## learning pandas methods 
- importing a csv file **pd.read_csv('file/path')**
- calling out the first rows   **head()**
- calling out the last rows    **tail()**
- calling out a range of rows  **iloc[:]**
- calling the dataframes summary of the percentiles, count, mean etc **describe()**
- shows the specific characters in a column **unique()**
- calling a specific character in a column **df[df['job']=='Data_analyst']**
- other metnods **& , notna()**
- sorting **sort_value() or sort_index()**
- deleting a column **drop()**
- grouping characters in a column **group_by()**
- **agg(['mean', 'median', 'max'])**
- **lambda X:**
- other methods **isin() , any()**
- ploting **plot(kind='barh')**
- **plt.title('name title')**
- **plt.ylabel('name ylable')**
- **plt.xlabel('name xlable')**
- **plt.xticks(rotation=45, ha = 'right')**
- **plt.show()**

##  Python data cleaning

```python
df.info()
```

```python
type(df.job_posted_date[0])
```

```python
# changing date type to datetime
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
# checking if the type for date has change to datetime
# it changed from object to datetime
df.info()
```

```python
type('job_posted_date')
```

```python
# dt. can now provide opportunities to change date in to day, month, year etc 

df.job_posted_date.dt.month
```

```python
# creating a new column for the month

df["job_posted_month"] = df.job_posted_date.dt.month
```

```python
# job_posted_month the new column has been added into the table

df.info()
```

```python
# sorting data
# Inplace makes it permanently sorted or you can also use df = df.sort_values('job_posted_date')
# you can also use df = df.sort_values(by = 'job_posted_date') either way works

df.sort_values('job_posted_date', inplace = True)
```

```python
df
```

```python
# removing a column
# label = mention the column name
# axis = (0 = rows, 1= columns) so this indictates what to drop, row or column

df.drop(labels ='salary_hour_avg', axis = 1, inplace=True)
```

```python
# Column removed
df.info()
```

```python
# removing the null values

df.dropna(subset='salary_year_avg', inplace=True)
```

```python
df.info()
```

### Panda analysis

```python
dn=pd.read_csv('C:/Users/Ian/Desktop/panda ex/data_jobs2.csv')
```

```python
dn.describe()
```

```python
# group by 
# eg groupby(['Animal']).mean() this groups the characters in a specified column
dn.groupby(['job_title_short'])['salary_year_avg'].median()
```

```python
# can also group more than one column
dn.groupby(['job_title_short', 'job_country'])['salary_year_avg'].median()
```

```python
# remember when making more than one column the have to be in a list "[]"
dn.groupby(['job_title_short'])[['salary_year_avg', 'salary_hour_avg']].median()
```

```python
# you want more columns of min, max, median etc use "agg"
# note: if the characters are more one make it a list "[]""
dn.groupby(['job_title_short'])[['salary_year_avg']].agg(['min', 'max', 'median'])
```

```python
dn.groupby(['job_title_short'])['salary_year_avg'].apply(lambda x: f'{x.min()} - {x.max()}')
```

```python
dn['job_posted_date']=pd.to_datetime(dn['job_posted_date'])
```

```python
us_jobs.groupby(['job_title_short'])['salary_year_avg'].agg(['count','min','max','median']).sort_values('median', ascending = False)
```

### ploting charts in pandas

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
job_counts = dn.job_title_short.value_counts()
job_counts.plot(kind="line")
```

```python
job_counts = dn.job_title_short.value_counts()
job_counts.plot(kind="bar")
plt.bar(job_counts.index, job_counts)
plt.title('Postings by job title')
plt.xlabel('')
plt.ylabel('count of job postings')
plt.xticks(rotation=45, ha = 'right')  # this was got from chatgpt. ha='right' means alignment of x lables to the right
plt.show()
```
